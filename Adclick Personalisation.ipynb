```python
df['City Codes']= df['City in UK'].astype('category').cat.codes
```

```python
df['Network']= df['network'].astype('category').cat.codes
```

```python
df['device_id']= df['device_id'].astype('category').cat.codes
```

```python
df['show_title']= df['show_title'].astype('category').cat.codes
```

```python
df['Month'] = df['viewing_start'].apply(lambda x: x.split('/')[1])
df['Hour'] = df['viewing_start'].apply(lambda x: x.split(':')[0].split(' ')[1])
```

**Training a Logistic Regression Model**

```python
X = df.drop(labels=['Ad topic line','City in UK', 'viewing_session_length','viewing_start','scheduled_start', 'network', 'ip_address', 'Ad on Click'], axis=1)
```

```python
X.head()
```

```python
y = df['Ad on Click']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

**Predictions and Evaluations**

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions, target_names=['Not Clicked','Clicked']))
```
